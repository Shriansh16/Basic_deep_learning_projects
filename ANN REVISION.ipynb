```python
import tensorflow as tf
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv('/content/Churn_Modelling.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.drop(columns=['RowNumber','CustomerId','Surname'],inplace=True)
```

```python
df.head()
```

```python
x=df.iloc[:,:-1]
```

```python
x
```

```python
y=df.iloc[:,-1]
```

```python
y
```

```python
x.head()
```

```python
x['Geography'].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python

from sklearn.compose import ColumnTransformer
```

```python
cat_col=x[['Geography','Gender']]
```

```python
ohe=OneHotEncoder()
```

```python
encoded_columns = ohe.fit_transform(cat_col)
```

```python
encoded_df = pd.DataFrame(encoded_columns.toarray(), columns=ohe.get_feature_names_out(['Geography','Gender']))
```

```python
encoded_df
```

```python
x.drop(columns=['Geography','Gender'],inplace=True)
```

```python
x=pd.concat([x,encoded_df],axis=1)
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# feature scaling
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
X_train
```

```python
# LETS CREATE ANN
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LeakyReLU,ELU,PReLU
from tensorflow.keras.layers import Dropout
```

```python
#lets initialize ANN
classifier=Sequential()
```

```python
#adding the input layer
classifier.add(Dense(units=11,activation='relu'))
```

```python
#adding the first hidden layer
classifier.add(Dense(units=7,activation='relu'))
#adding dropout layer
#classifier.add(Dropout(0.2))
```

```python
#adding second hidden layer
classifier.add(Dense(units=6,activation='relu'))
```

```python
#adding output layer
classifier.add(Dense(units=1,activation='sigmoid'))
```

```python
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model_history=classifier.fit(X_train,y_train,validation_split=0.33,batch_size=10,epochs=10)
```

```python
model_history.history.keys()
```

```python
y_pred=classifier.predict(X_test)
y_pred=(y_pred>=0.5)
```

```python
#make confusion metrics
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
cm=confusion_matrix(y_test,y_pred)
```

```python
cm
```

```python
accuracy_score(y_test,y_pred)
```
